```r
# Problem 6

install.packages("ggmap",repos = "http://cran.us.r-project.org")
install.packages("maptools",repos = "http://cran.us.r-project.org")
install.packages("lattice",repos = "http://cran.us.r-project.org")

library("ggmap")
library("maptools")
library(maps)
visited <- c( "Delhi", "London", "Chicago", "SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
visited <- c("Denver", "Houston", "California")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit.x,visit.y, col="yellow", pch=36)
```
